```python
import pandas as pd
```

```python
# read in the excel file into a pandas dataframe
df = pd.read_excel("duplicate_records_toy.xlsx")

# show df
df
```

```python
# find all duplicated values based on patient_id
df_duplicated = df[df.duplicated(subset = "patient_id", keep = False)]

# show df_duplicated
df_duplicated
```

```python
# drop complete duplicates
df.drop_duplicates(inplace = True)

# show df
df
```

```python
# correct misspellings

# locate relevant patient
condition = df['patient_id'] == 10002

# change last name
df.loc[condition, 'last_name'] = 'McGuire'

# show df
df
```

```python
# drop complete duplicates
df.drop_duplicates(inplace = True)

# show df
df
```

```python
# drop duplicate recrods, keep first
df.drop_duplicates(subset = 'patient_id', keep = 'last', inplace = True)

# sort df desc by patient_id
df = df.sort_values(by='patient_id')

# show df
df
```
